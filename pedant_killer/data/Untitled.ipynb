```python
import pandas as pd
```

```python
df_device_service = pd.read_csv('device_service.csv')
```

```python
new_rows = []

for device, group in df_device_service.groupby("Устройство", sort=False):
    new_rows.append(group)  # добавляем все строки устройства

    manufacturer = group["Производитель"].iloc[0]
    device_type = group["Тип девайса"].iloc[0]

    # Создаём строку диагностики
    diag_row = pd.DataFrame([{
        "Устройство": device,
        "Услуга": "Диагностика",
        "Цена": 0.0,
        "Гарантия": 0.0,
        "Производитель": manufacturer,
        "Тип девайса": device_type
    }])

    new_rows.append(diag_row)  # вставляем диагностику сразу после

# Объединяем всё в итоговый DataFrame
df_ordered = pd.concat(new_rows, ignore_index=True)

# Сохраняем результат
df_ordered.to_csv("устройства_с_диагностикой.csv", index=False)
```
